### DB-Scan Clustering
- Points
    - Core Points
        - there should be atleast minimum points should present in drawn circle with epsilon radius.
    - border points
        - points which are not core points but are present in the circle drawn with epsilon radius around core
        - no of points are less than the minimum points with epsilon radius
    - outliers
        - No other points present with epsilon radius. 
    - minimum points , radius(epsilon) of circle
- able to handle the Outliers.
- able to handle non-linear clustering.
- Pros and Cons


| Pros|Cons |
|--|--|
|No need to specify no of Clusters|not entirelt deterministic|
|can find arbitarily- shapde Clusters|depends on distance measure used|
|NOtion of Noise and robust to Outliers|can not cluster datasets with larger differnce in densities| 
|designed for use with database that can accelerate region quiries|if data and scale are not well understood,choosing params in difficult|
|the params minPts and epsilon can be set by a domain expert|


```python
### DBScan implementation
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
```

```python
X,y = make_moons(n_samples=250,noise=0.05)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
### Feature Scaling
from sklearn.preprocessing import StandardScaler
scaler  = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
dbscan = DBSCAN(eps=0.3)
dbscan.fit(X_scaled)
```

```python
dbscan.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c=dbscan.labels_)
```
