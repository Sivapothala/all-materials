```python
import matplotlib.pyplot as plt
```

```python
x =[1,2,3,4,5]
y=[4,5,6,7,8]

# plt.plot(x,y)
plt.plot(x,y,color='red',linestyle='-.')
```

```python
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]

plt.figure(figsize=(9,5))
plt.subplot(1,2,1)
plt.plot(x,y1,color='green')
plt.title("plot 1")

plt.subplot(1,2,2)
plt.plot(x,y2,color='red')
plt.title("plot 2")
```

```python
cat = ['a','b','c']
val = [10,20,30]

plt.bar(cat,val,color='purple')
plt.xlabel('Categories')
plt.ylabel('values')
plt.title('Bar plots')
plt.show()
```

#### Histograms
histograms are used to represent the distribution of a dataset. They divide the data into bins and count the number of points in each bin.

```python
data = [1,2,2,2,3,1,5,4,4,4,6,6,6,8]
##create a histogram
plt.hist(data,bins=5,color='orange',edgecolor='black')
```

```python
# create a scatter plot
plt.scatter(x,y1,marker='x')
```

```python
#create a pie chart
labels = ['a','b','c','d']
sizes = [30,20,40,10]
color = ['green','yellow','orange','blue']
explode=(0.2,0,0,0)

plt.pie(sizes,explode=explode,labels=labels,colors=color,autopct="%1.1f%%",shadow=True)
```
